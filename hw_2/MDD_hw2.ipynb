```python
##CODE TO MAKE PLOT IN BOKEH
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import Range1d

nitrate = np.loadtxt('ni_II_nitrate_fin.txt',delimiter=',') 
trisbipy =  np.loadtxt('ni_II_trisbipy_fin.txt', delimiter=',') 
mcr =  np.loadtxt('MCR_ox_fin.txt',  delimiter=',') 


p = figure(title="", x_axis_label='Energy (eV)', y_axis_label='Delta Intensity (a.u.)           Normalized Intensity (a.u.)')

p.line(mcr[0], mcr[1], legend="MCR", color='blue', line_width=2)
p.line(nitrate[0], nitrate[1], legend="Ni(II)-hexaquo", color='orange',line_width=2)
p.line(trisbipy[0], trisbipy[1], legend=" Ni(II)-trisbipy",color='green', line_width=2)

p.line(trisbipy[0], nitrate[1] - trisbipy[1] - 0.2, color='red', legend="Ni(II)-hexaquo - Ni(II)-hexaquo", line_width=2)
p.line(trisbipy[0], mcr[1] - nitrate[1]  - 0.4, color='purple', legend="MCR - Ni(II)-hexaquo", line_width=2)
p.line(trisbipy[0], mcr[1] - trisbipy[1] - 0.6 ,color='brown', legend="MCR - Ni(II)-trisbipy",line_width=2)
p.legend.location = "top_left"
p.y_range = Range1d(-1, 1.3)
show(p)
```

a) The CSV file `data/california_pressure_2022_01_15.csv` contains [Automated Surface Observing System (ASOS)](https://mesonet.agron.iastate.edu/ASOS/) pressure measurements at 64 stations in California taken several hours later that day, observed with 1-minute resolution. Read this data into a `pandas` DataFrame and then plot in the same figure the pressure as a function of time at SBA (Santa Barbara Municipal Airport, in Southern California), OAK (Oakland Airport, Bay Area), and MHS (Mount Shasta City, in Northern California). To allow a better visual comparison between the three, divide each plot by the initial value at the start of each of the timeseries. Overplot a 5-min thick vertical line for each timeseries (with the same color as the line) at the maximum of that series.

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import dateutil.parser as parser
shock = pd.read_csv('data/california_pressure_2022_01_15.csv')
shock.rename(columns = {'valid(UTC)':'Time'}, inplace = True)
shock2=shock.copy()
shock['Time'] = pd.to_datetime(shock['Time'])
shock.set_index('Time', inplace=True)

plt.figure(figsize=[10,5])
plt.plot(shock['SBA']/shock['SBA'][0], label='SBA', color='blue')
plt.plot(shock['OAK']/shock['OAK'][0],label='OAK', color='orange')
plt.plot(shock['MHS']/shock['MHS'][0],label='MHS', color='green')


sba_max=shock2['SBA'].idxmax()
oak_max = shock2['OAK'].idxmax()
mhs_max =shock2['MHS'].idxmax()

plt.vlines(shock2['Time'][sba_max-2:sba_max+3],ymin=0.999, ymax=1.0035,color='blue')
plt.vlines(shock2['Time'][oak_max-2:oak_max+3],ymin=0.999, ymax=1.0035,color='orange')
plt.vlines(shock2['Time'][mhs_max-2:mhs_max+3],ymin=0.999, ymax=1.0035,color='green')
plt.xlabel('Date')
plt.ylabel('Pressure (Normalized)')
plt.legend()
plt.show()

print ('Max for SBA = ' , shock['SBA'].idxmax())
print ('Max for OAK = ',  shock['OAK'].idxmax())
print ('Max for MHS = ',  shock['MHS'].idxmax())
```
